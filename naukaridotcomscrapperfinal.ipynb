```python
# load library
import csv
import time
from selenium import webdriver
import pandas as pd
from bs4 import BeautifulSoup
```

# '''First import all necessory modules imported'''

```python
# chromedriver url link
driver=webdriver.Chrome()
```

# '''Below is I am creating dictionary to store data which later conerted to data frame'''

```python
# create dictionary
jobs={"company":[],
    "discription":[],
      "exp":[],
     "loc":[],
     "salary":[],
     "skills":[]}
```

# ''' below is main code for scrapping data from naukari.com from pages upto 20 and then data append to Dictionary created above '''

```python
# link of website you want to scrap
for i in range(1,21):
    driver.get("https://www.naukri.com/software-developer-jobs".format(i))
    time.sleep(3)
    lst=driver.find_elements_by_css_selector(".jobTuple.bgWhite.br4.mb-8")
     # scrape the data from website 
    for job in lst:
        driver.implicitly_wait(10)
        role=job.find_element_by_css_selector("a.title.fw500.ellipsis").text
        company=job.find_element_by_css_selector("a.subTitle.ellipsis.fleft").text
        location=job.find_element_by_css_selector(".fleft.grey-text.br2.placeHolderLi.location").text
        exp=job.find_element_by_css_selector(".fleft.grey-text.br2.placeHolderLi.experience").text
        salaries=job.find_element_by_css_selector(".fleft.grey-text.br2.placeHolderLi.salary").text
        skills=job.find_element_by_css_selector(".tags.has-description").text
        jobs["company"].append(company)
        jobs["discription"].append(role)
        jobs["exp"].append(exp)
        jobs["loc"].append(location)
        jobs["salary"].append(salaries)
        jobs["skills"].append(skills)
        
```

# '''below print data collected in dictionary'''

```python
# print scrape data
print(jobs)
```

# '''Data from dictionary jobs converted to df dataframe'''

```python
# tranform data into dataframe
df=pd.DataFrame(jobs)
```
